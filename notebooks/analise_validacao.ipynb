```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Simulação dos dados finais do ETL
data = {
    'brick': [
        '1147 - CAMPO GRANDE - CENTRO',
        '1147 - CAMPO GRANDE - CENTRO',
        '1147 - CAMPO GRANDE - CENTRO',
        '1147 - CAMPO GRANDE - CENTRO',
        '1147 - CAMPO GRANDE - CENTRO'
    ],
    'ean': ['32689150', '42110200', '42176763', '42277217', '42355014'],
    'vol_pp': [5.0, 5.0, 0.0, 23.0, 1.0],
    'vol_total_mercado': [5.0, 12.0, 4.0, 115.0, 3.0]
}

df_final = pd.DataFrame(data)
df_final
```

```python
df_grouped = df_final.groupby('brick')[['vol_pp', 'vol_total_mercado']].sum()

df_grouped.plot(kind='bar')
plt.title('Volume PP vs Volume Total de Mercado por Brick')
plt.xlabel('Brick')
plt.ylabel('Volume')
plt.show()
```

```python
df_final[df_final['ean'] == '32689150']['vol_pp'].hist()
plt.title('Distribuição de vendas do produto EAN 32689150')
plt.xlabel('Volume PP')
plt.ylabel('Frequência')
plt.show()
```
